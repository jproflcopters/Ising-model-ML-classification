```python
import numpy as np
import pickle

blok0 = np.zeros((10000,1), dtype='int')
blok1 = np.ones((10000,1), dtype='int')
blok2 = 2*np.ones((10000,1), dtype='int')
blok = np.vstack((blok0, blok1, blok2))

np.save("dataset", blok)
```

```python
print(blok0.shape, blok1.shape, blok2.shape)
```

```python
blok.shape
```

```python
name='Triple.pkl'
pickle.dump(np.packbits(blok), open(name,'wb'))
```
