```python
import numpy as np
import pickle

blok0 = np.zeros((10000,1), dtype='int')
blok1 = np.ones((10000,1), dtype='int')
blok2 = 2*np.ones((10000,1), dtype='int')
blok = np.vstack((blok0, blok1, blok2))

print('Shapes of blok0, blok1, blok2, blok:', str(blok0.shape)
      +', '+str(blok1.shape)+', '+str(blok2.shape)+', '+str(blok.shape))
```

```python
np.save("Triple", blok) # use this for 3 and more labels (0, 1, 2)
print('Saved.')
```

```python
name='Two_labels.pkl'
pickle.dump(np.packbits(blok), open(name,'wb')) # use that for up to two labels (0, 1)
print('Saved.')
```
